```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

def create_random_value(mean, std, minimum, maximum):
    value = np.random.normal(mean, std)
    return min(maximum, max(minimum, value))

def dataset_generator(avg, max_vals, min_vals, std, start_time, minutes):
    columns = ['저압보일러: 급수유량','반응기1온도: 입구','반응기1온도: 출구','HE1온도: 입구', '반응기2온도: 입구', '반응기2온도: 출구', '콘덴서: 압력', '콘덴서: 레벨', '콘덴서: 급수유량', '유입량: 순수유입량', 'AcidGas: 유량','AcidGas: 압력', 'AcidGas: 온도', 'COR(촉매반응): 1차 Air',
           'COR(촉매반응): 2차 Air','COR(촉매반응): Air 압력','COR(촉매반응): COG 유량','COR(촉매반응): 상부온도','COR(촉매반응): 중부온도', 'COR(촉매반응): 하부온도', '탈황고압보일러: 출구온도', '탈황고압보일러: 스팀유량', '탈황고압보일러: 압력', '탈황고압보일러: 급수유량', 
           '탈황고압보일러: 레벨', 'PG Heat 온도: 1출구', 'TailGas: 유량', 'TailGas: 온도', 'DrainTank: 레벨', 'DrainTank: 온도','저장Tank: 온도', 'DesuperHeat: 압력', 'DesuperHeat: 레벨', 'TailGas분석기: H2S', 'TailGas분석기: SO2' ]
    data = []

    current_time = start_time
    for _ in range(minutes):
        row = [create_random_value(avg[i], std[i], min_vals[i], max_vals[i]) for i in range(len(columns))]
        data.append([current_time] + row)
        current_time += timedelta(minutes=1)

    df = pd.DataFrame(data, columns=['time_idx'] + columns)
    df.set_index('time_idx', inplace=True)
    return df
```

```python
start_time = datetime.now()
generated_data = dataset_generator(average, maximum, minimum, std, start_time, 60 * 24 * 365)  # Generate data for 60 minutes
# 1년치 데이터 저장 

# 데이터 저장
with open('./data/CO2/raw/co2.pkl', 'wb') as f:
	pickle.dump(generated_data, f)
```
